```python
# Create squad table
squad_df = events_df[['player_id', 'player', 'position', 'team']].drop_duplicates()

# Filter to only include Bayer players
squad_bayer = squad_df[squad_df['team'] == 'Bayer Leverkusen']
```

```python
# Remove rows with unidentified players
squad_bayer_clean = squad_bayer.dropna(subset=['player_id', 'player'])

# Players may have multiple positions throughout the season
# Count the number of occurrences of each position per player
position_counts = squad_bayer_clean.groupby(['player_id', 'player', 'position']).size().reset_index(name='count')

# Keep the most frequent position for each player
bayer_squad = position_counts.sort_values(['player_id', 'count'], ascending=[True, False]).drop_duplicates(subset='player_id')

# Create a jersey_name column by taking the last word of the full name
bayer_squad['common_name'] = bayer_squad['player'].apply(lambda x: x.split()[-1])

# Dictionary for players with multiple names
name_corrections = {
    'Alejandro Grimaldo García': 'Alejandro Grimaldo',
    'Borja Iglesias Quintas': 'Borja Iglesias',
    'Exequiel Alejandro Palacios': 'Exequiel Palacios',
    'Edmond Fayçal Tapsoba': 'Edmond Tapsoba',
    'Victor Okoh Boniface': 'Victor Boniface',
    'Piero Martín Hincapié Reyna': 'Piero Hincapié',
    'Arthur Augusto de Matos Soares': 'Arthur',
    'Gustavo Adolfo Puerta Molano': 'Gustavo Puerta'
}

# Apply the corrections
def generate_common_name(name):
    return name_corrections.get(name, name)

# Create the new column
bayer_squad['common_name'] = bayer_squad['player'].apply(generate_common_name)
bayer_squad.drop('count', axis=1, inplace=True)
```

```python
bayer_squad_list = []

for match_id in bayer_match_ids:
    # Get Bayer lineups for the match
    lineup_df = sb.lineups(match_id=match_id)[team]
    
    # Select relevant columns
    lineup_df = lineup_df[['player_id', 'player_name', 'jersey_number', 'country']]
    
    bayer_squad_list.append(lineup_df)

# Concatenate all lineups
bayer_squad_df = pd.concat(bayer_squad_list, ignore_index=True)

# Remove duplicates (keep one row per player_id)
bayer_squad_df = bayer_squad_df.drop_duplicates(subset='player_id').reset_index(drop=True)

# Name corrections
name_corrections = {
    'Alejandro Grimaldo García': 'Alejandro Grimaldo',
    'Borja Iglesias Quintas': 'Borja Iglesias',
    'Exequiel Alejandro Palacios': 'Exequiel Palacios',
    'Edmond Fayçal Tapsoba': 'Edmond Tapsoba',
    'Victor Okoh Boniface': 'Victor Boniface',
    'Piero Martín Hincapié Reyna': 'Piero Hincapié',
    'Arthur Augusto de Matos Soares': 'Arthur',
    'Gustavo Adolfo Puerta Molano': 'Gustavo Puerta'
}

def apply_common_name(name):
    return name_corrections.get(name, name)

bayer_squad_df['common_name'] = bayer_squad_df['player_name'].apply(apply_common_name)

def get_second_or_first_word_upper(name):
    if isinstance(name, str):
        parts = name.strip().split()
        if len(parts) > 1:
            return parts[1].upper()
        elif parts:
            return parts[0].upper()
    return None

bayer_squad_df['jersey_name'] = bayer_squad_df['common_name'].apply(get_second_or_first_word_upper)

# Keep only players who have recorded events
players_with_event = events_df['player_id'].dropna().unique()
bayer_squad_df = bayer_squad_df[bayer_squad_df['player_id'].isin(players_with_event)]

bayer_squad_df
```

## Composition la plus utilisée

```python
# Get all Bayer lineups (already done)
lineups = []

for match_id in bayer_match_ids:
    df = sb.lineups(match_id)[team]
    df['match_id'] = match_id
    lineups.append(df)

lineups_df = pd.concat(lineups, ignore_index=True)

print("Available columns:", lineups_df.columns.tolist())
```

```python
lineups_df
```

```python
# Explode the list of positions
exploded = lineups_df.explode('positions', ignore_index=True)

# Convert the column of dicts into separate columns
positions_df = pd.json_normalize(exploded['positions'])

# Concatenate with the rest of the data
exploded = pd.concat([exploded.drop(columns='positions'), positions_df], axis=1)

# Keep only the starters
starters_df = exploded[exploded['start_reason'] == 'Starting XI']

# Count the number of starts by player and position
starts_by_pos = (
    starters_df.groupby(['player_id', 'position'])
    .size()
    .reset_index(name='starts')
)

# Keep each player's most frequent position
main_position = (
    starts_by_pos.sort_values('starts', ascending=False)
    .drop_duplicates('player_id')
)

# Get player info (name, country, etc.)
player_info = lineups_df.drop_duplicates('player_id')[['player_id', 'player_name', 'jersey_number', 'country']]

# Merge with main position data
final_df = main_position.merge(player_info, on='player_id', how='left')

# Keep the 11 players with the most starts
starting_xi = final_df.sort_values('starts', ascending=False).head(11)

# Reorder columns
starting_xi = starting_xi[['player_id', 'jersey_number', 'player_name', 'country', 'position', 'starts']]
```

```python
# Explode positions
exploded = lineups_df.explode('positions', ignore_index=True)

# Convert dicts into separate columns
positions_df = pd.json_normalize(exploded['positions'])
exploded = pd.concat([exploded.drop(columns='positions'), positions_df], axis=1)

# Keep starters only
starters_df = exploded[exploded['start_reason'] == 'Starting XI']

# Count starts by player and position
starts_by_position = (
    starters_df.groupby(['player_id', 'position'])
    .size()
    .reset_index(name='starts')
)

# Keep each player's main position (most played)
main_position = (
    starts_by_position.sort_values('starts', ascending=False)
    .drop_duplicates('player_id')
    .rename(columns={'position': 'main_position'})
)

# Count total starts across all positions
total_starts = (
    starters_df.groupby('player_id')
    .size()
    .reset_index(name='starts')
)

# Player info
player_info = lineups_df.drop_duplicates('player_id')[['player_id', 'player_name', 'jersey_number', 'country']]

# Final merge
xi_df = total_starts.merge(player_info, on='player_id', how='left')
xi_df = xi_df.merge(main_position[['player_id', 'main_position']], on='player_id', how='left')

# Sort and keep top 11 most used players
xi_df = xi_df.sort_values('starts', ascending=False).head(11)

# Reorder columns
xi_df = xi_df[['player_id', 'jersey_number', 'player_name', 'country', 'main_position', 'starts']]
```

```python
# Coordinate dictionary
position_coords = {
    'Goalkeeper': (50, 5),  # At the bottom

    'Left Center Back': (30, 20),
    'Center Back': (50, 20),
    'Right Center Back': (70, 20),

    'Left Wing Back': (15, 35),
    'Right Wing Back': (85, 35),

    'Left Defensive Midfield': (35, 33),
    'Right Defensive Midfield': (65, 33),

    'Left Attacking Midfield': (35, 45),
    'Right Attacking Midfield': (65, 45),

    'Center Forward': (50, 55)  # At the top
}

# Add x and y columns based on position
xi_df['x'] = xi_df['main_position'].map(lambda pos: position_coords.get(pos, (None, None))[0])
xi_df['y'] = xi_df['main_position'].map(lambda pos: position_coords.get(pos, (None, None))[1])
```

## Minutes jouées

```python
# Explode positions
exploded = lineups_df.explode('positions', ignore_index=True)

# Convert to columns
positions_df = pd.json_normalize(exploded['positions'])
exploded = pd.concat([exploded.drop(columns='positions'), positions_df], axis=1)

# Convert time strings to seconds
def time_to_seconds_from(t):
    if pd.isna(t):
        return 0
    m, s = map(int, t.split(':'))
    return m * 60 + s

def time_to_seconds_to(t):
    if pd.isna(t):
        return 90 * 60  # player stayed until the end
    m, s = map(int, t.split(':'))
    return m * 60 + s

exploded['from_sec'] = exploded['from'].apply(time_to_seconds_from)
exploded['to_sec'] = exploded['to'].apply(time_to_seconds_to)

# Calculate duration in minutes
exploded['duration_min'] = (exploded['to_sec'] - exploded['from_sec']) / 60

# Just remove any potential negative values
exploded = exploded[exploded['duration_min'] >= 0]

# Aggregate by player and match
minutes_df = (
    exploded.groupby(['match_id', 'player_id', 'player_name'])
    .agg({'duration_min': 'sum'})
    .reset_index()
    .rename(columns={'duration_min': 'minutes_played'})
)

minutes_df
```
